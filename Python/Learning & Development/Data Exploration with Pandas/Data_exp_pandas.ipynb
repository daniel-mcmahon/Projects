```python
# College with highest mid-career salary
clean_df['Mid-Career Median Salary'].idxmax()
clean_df.loc[8]

# College Major = Chemical Engineering
# Starting Grads Earn = 63200
```

```python
# College Major with the lowest starting salary
clean_df['Starting Median Salary'].idxmin()
clean_df.loc[49]

# College Major with lowest start salary = Spanish
# Grads after uni earn = 34000
```

```python
# College major with lowest mid career salary
clean_df['Mid-Career Median Salary'].idxmin()
clean_df.loc[18]

# College major with the lowest mid career salary is Education
# Mid Career Salary is 52000
```

```python
# Calculate the spread of 10th and 90th percentile

spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

```python
# Sort by low risk to high

low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
# Challenge - Find the 5 degrees with the highest potential

most_potential = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
most_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
# Find the degrees with the greatest spread in Salaries

high_risk = clean_df.sort_values('Spread', ascending=False)
high_risk[['Undergraduate Major', 'Spread']].head()
```

```python
# Using the group by function

pd.options.display.float_format = '{:,.2f}'.format 
clean_df.groupby('Group').mean()
```
